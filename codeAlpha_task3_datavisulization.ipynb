```python
import pandas as pd

data = {
    "Name": ["Harsh", "Rahul", "Aman", "Riya", "Priya", "Karan", "Sneha", "Ankit", "Pooja", "Neha"],
    "Math": [85, 78, 92, 88, 76, 81, 95, 70, 89, 83],
    "Science": [90, 80, 95, 84, 82, 77, 94, 73, 91, 85],
    "English": [88, 75, 91, 90, 79, 80, 96, 72, 90, 84]
}

df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Math"])
plt.title("Math Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.savefig("BarChart.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Name"], df["Science"], marker="o")
plt.title("Science Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.savefig("LineChart.png")
plt.show()
```

```python
total = [df["Math"].sum(), df["Science"].sum(), df["English"].sum()]

subjects = ["Math", "Science", "English"]

plt.figure(figsize=(6,6))
plt.pie(total, labels=subjects, autopct="%1.1f%%")
plt.title("Subject Wise Total Marks")
plt.savefig("PieChart.png")
plt.show()
```
