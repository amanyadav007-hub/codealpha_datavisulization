```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

print("Web Scraping libraries successfully imported!")
```

```python
# Target URL
url = "https://quotes.toscrape.com/"

# Sending an HTTP GET request to fetch the HTML content of the website
response = requests.get(url)

# Empty list to store the scraped data
scraped_data = []

if response.status_code == 200:
    # Parsing the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Target every individual quote container block
    quotes = soup.find_all('div', class_='quote')

    for q in quotes:
        # Extracting text, author, and tags from each quote container
        text = q.find('span', class_='text').text.replace('“', '').replace('”', '')
        author = q.find('small', class_='author').text
        tags = [tag.text for tag in q.find_all('a', class_='tag')]

        # Storing the extracted details in a structured dictionary format
        scraped_data.append({
            'Quote': text,
            'Author': author,
            'Tags': ', '.join(tags)
        })
    print(f"Successfully scraped {len(scraped_data)} rows of data!")
else:
   print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python

# Convert the data into a Pandas DataFrame
df_scraped = pd.DataFrame(scraped_data)

# Display the first 5 rows to check
print("=== Scraped Dataset Preview ===")
display(df_scraped.head())

# Save the data as a CSV file in the local environment
df_scraped.to_csv("scraped_quotes_dataset.csv", index=False)
print("\nDataset saved successfully as 'scraped_quotes_dataset.csv'!")
```
