```python
# Essential Libraries for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings to make visualizations beautiful and clean
sns.set_theme(style="darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)
print("Libraries successfully imported!")
```

```python
!unzip -o "archive (2).zip"
import pandas as pd

df = pd.read_csv("titanic.csv")
print("Titanic dataset successfully loaded!")

# Display the first 5 rows of the dataset
print("\nFirst 5 rows of the dataset:")
display(df.head())
```

```python
# If the Age column has missing values, fill them with the median
if df['Age'].isnull().sum() > 0:
    df['Age'] = df['Age'].fillna(df['Age'].median())
    print("Missing values in Age column have been filled with the median.")

# View the statistical overview of the data
df.describe().T
```

```python
# 1. Fare Distribution Chart (Histplot)
plt.figure(figsize=(8, 4))
sns.histplot(df['Fare'], kde=True, color='teal')
plt.title('Passenger Fare Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

# # 2. Survival Count by Passenger Class (Countplot)
plt.figure(figsize=(8, 4))
# Filtering the dataframe to only include passengers who survived (Survived == 1)
sns.countplot(data=df[df['Survived'] == 1], x='Pclass', palette='muted', hue='Pclass', legend=False)
plt.title('Total Survivors by Passenger Class', fontsize=14, fontweight='bold')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')
plt.show()

# 3. Correlation Matrix (Heatmap)
plt.figure(figsize=(6, 4))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap', fontsize=14, fontweight='bold')
plt.show()
```
